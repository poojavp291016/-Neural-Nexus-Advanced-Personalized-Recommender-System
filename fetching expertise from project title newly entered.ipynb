```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Define words to remove
words_to_remove = ["and", "of", "for", "on", "its", "the"]

# Define technical keywords
technical = ["development", "engineering", "management", "science"]

# Combine words to remove and technical keywords
words_to_remove.extend(technical)

def remove_stopwords_and_extract_keywords(text: str, words_to_remove: list):
    # Remove stopwords
    text = " ".join([word for word in text.split() if word.lower() not in words_to_remove])
    # Extract keywords
    doc = nlp(text)
    keywords = [token.text for token in doc if token.pos_ in ["NOUN", "PROPN"]]
    return keywords

def spacy_test(project_title: str, expertise_list: list, threshold=0.8):
    # Remove stopwords and extract keywords from project title
    project_keywords = remove_stopwords_and_extract_keywords(project_title, words_to_remove)

    # Iterate through expertise list
    matching_expertise = []
    for expertise in expertise_list:
        expertise_keywords = remove_stopwords_and_extract_keywords(expertise, words_to_remove)
        common_keywords = set(project_keywords).intersection(set(expertise_keywords))
        similarity = len(common_keywords) / max(len(project_keywords), len(expertise_keywords))
        print(f"Comparing project title '{project_title}' with expertise '{expertise}':")
        print(f"Common keywords: {common_keywords}")
        print(f"Similarity: {similarity}")
        if similarity > threshold:
            matching_expertise.append(expertise)
    
    return matching_expertise

# Example usage:
project_title = input("Enter project title: ")

# Load expertise list from file
with open("C:\\Users\\pooja\\expertise.csv", 'r') as file:
    expertise_list = file.readlines()

# Find matching expertise
matching_expertise = spacy_test(project_title, expertise_list)
print("Matching expertise:")
for expertise in matching_expertise:
    print("- ", expertise.strip())
```
