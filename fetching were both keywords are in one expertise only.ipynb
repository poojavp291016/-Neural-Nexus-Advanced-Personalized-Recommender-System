```python
import pandas as pd

# Function to find matching expertise based on project keywords
def find_matching_expertise(project_keywords, expertise_list):
    matched_expertise = []

    # Convert project keywords to lowercase for case-insensitive matching
    project_keywords = [keyword.lower() for keyword in project_keywords]

    # Iterate over expertise list and check for presence of all project keywords
    for expertise in expertise_list:
        expertise_words = expertise.lower().split()
        if all(keyword in expertise_words for keyword in project_keywords):
            matched_expertise.append(expertise)

    return matched_expertise

# Specify the file path for expertise CSV
expertise_file = r"C:\Users\pooja\expertise.csv"

# Load and inspect the contents of the CSV file
try:
    df = pd.read_csv(expertise_file)
    print("Contents of the CSV file:")
    print(df)  # Display the DataFrame contents
    print()

    # Extract expertise descriptions from the DataFrame
    expertise_list = df['name'].tolist()

    # Example usage: Get project title from user input
    project_title = input("Enter project title: ")

    # Split project title into individual keywords
    project_keywords = project_title.lower().split()

    # Find matching expertise based on the project keywords
    matched_expertise = find_matching_expertise(project_keywords, expertise_list)

    # Print the matched expertise
    if matched_expertise:
        print("\nMatched Expertise:")
        for expertise in matched_expertise:
            print("-", expertise)
    else:
        print("\nNo matching expertise found.")

except FileNotFoundError:
    print(f"Error: File not found at {expertise_file}")
except Exception as e:
    print(f"Error: {e}")
```
